```python
#Required imports
import requests
from tabulate import tabulate 
```

```python
#Needed environment variables
API_KEY = ""
Base_URL = "https://api.openweathermap.org/data/2.5/weather"
```

```python
#Conditional loop for inputs
while True:
    print("Do you want to find weather by coordinates or city?")
    #users can use coordinates or city names
    method = input("Enter 'coordinates' or 'city' ").lower()
    if method == 'city':  
        city = input("Enter city name: ")
        print(city)
        break
    elif method == 'coordinates':
        latitude = input("Enter latitude: ")
        longitude = input("Enter longitude: ")
        print(latitude)
        print(longitude)
        break
    else:
        print("Invalid option")
```

```python
#Function for calling with city name
def city_call(city, url, API_KEY):
    params = {
        "q" : city,
        "appid" : API_KEY,
        "units" : "imperial"
    }
    response = requests.get(url, params=params)
    return response
```

```python
#Function for calling with latitude and longitude
def coordinate_call(latitude, longitude, url, API_KEY):
    params = {
        "lat" : latitude,
        "lon" : longitude,
        "appid" : API_KEY,
        "units" : "imperial"
    }
    response = requests.get(url, params=params)
    return response
```

```python
#call correct function for the user's input
if method == 'city':
    response = city_call(city, Base_URL, API_KEY)
else:
    response = coordinate_call(latitude, longitude, Base_URL, API_KEY)
#print json. This proves the call worked
print(response)
data = response.json()
print(data)
```

```python
#Function to create a tabulate table for weather
def create_table(data):
    #weather table creates lists for important user features
    table = [
        ["City", data["name"]],
        ["Temperature (°F)", data["main"]["temp"]],
        ["Feels Like", data["main"]["feels_like"]],
        ["Humidity (%)", data["main"]["humidity"]],
        ["Wind (m/s)", data["wind"]["speed"]],
        ["Condition", data["weather"][0]["description"]],
    ]
    #Format table with tabulate
    display = tabulate(table, tablefmt="grid")
    return display
```

```python
#call creation and print table
table = create_table(data)
print(table)
```
